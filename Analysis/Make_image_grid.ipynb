```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import h5py
import random
from matplotlib.transforms import offset_copy
from matplotlib import rc
from matplotlib import patches
```

```python
freq  = [1, 3, 5]
cont = [-3, -6, -9, -12]

```

```python
for f in freq:
    for c in cont:
        file_name = '/tethys/kro18/MachineLearning_FinalProject/'+ str(f)+'MHz/'+ str(np.abs(c))+'dB/Lesion_Images/set_1_lesion_bmode.mat'
        
        file = h5py.File(file_name, 'r')
        img_data = file.get('bmode_img')
        img_data = np.array(img_data)
        
        img_num = random.randint(0,49)
        img = img_data[img_num, :, :]
        
        file_name = '/tethys/kro18/MachineLearning_FinalProject/'+ str(f)+'MHz/'+ str(np.abs(c))+'dB/Lesion_Images/set_1_lesion_coords.mat'
        file = h5py.File(file_name, 'r')
        coords_data = file.get('coords')
        coords_data = np.array(coords_data)
        
        cc = coords_data[img_num, :, :]
        
        
```

```python
rc('text', usetex=True)
rc('font', family='serif')
cols = ['{} dB'.format(col) for col in cont]
rows = ['{} MHz'.format(row) for row in freq]

fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12, 8))
plt.setp(axes.flat, xlabel='Lateral (cm)', ylabel='Axial (cm)')
ytck_lbs = np.arange(-1, 1, 0.5)
plt.setp(axes.flat, yticks=np.arange(0,390,97.25), yticklabels=np.arange(1, 3.1, 0.5))
plt.setp(axes.flat, xticks=np.arange(0,390,97.25), xticklabels=np.arange(-1, 1.1, 0.5))
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
#plt.rc('xlabel', fontsize=12)
#plt.rc('ylabel', fontsize=12)
pad = 5 # in points

for ax, col in zip(axes[0], cols):
    ax.annotate(col, xy=(0.5, 1), xytext=(0, pad),
                xycoords='axes fraction', textcoords='offset points',
                size=20, ha='center', va='baseline')

for ax, row in zip(axes[:,0], rows):
    ax.annotate(row, xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - pad, 0),
                xycoords=ax.yaxis.label, textcoords='offset points',
                size=20, ha='right', va='center')

fig.tight_layout()
# tight_layout doesn't take these labels into account. We'll need 
# to make some room. These numbers are are manually tweaked. 
# You could automatically calculate them, but it's a pain.
count=0
img_nums = np.zeros(12)
fig.subplots_adjust(left=0.15, top=0.95)
for i in range(len(freq)):
    for j in range(len(cont)):
        
        file_name = '/tethys/kro18/MachineLearning_FinalProject/'+ str(freq[i])+'MHz/'+ str(np.abs(cont[j]))+'dB/Lesion_Images/set_1_lesion_bmode.mat'
        
        file = h5py.File(file_name, 'r')
        img_data = file.get('bmode_img')
        img_data = np.array(img_data)
        
        img_num = random.randint(0,49)
        img_nums[count] = img_num
        img = img_data[img_num, :, :]
        
        file_name = '/tethys/kro18/MachineLearning_FinalProject/'+ str(freq[i])+'MHz/'+ str(np.abs(cont[j]))+'dB/Lesion_Images/set_1_lesion_coords.mat'
        file = h5py.File(file_name, 'r')
        coords_data = file.get('coords')
        coords_data = np.array(coords_data)
        
        cc = coords_data[img_num, :, :]
        
        axes[i][j].imshow(img, cmap='gray')
        #axes[i][j].xaxis.label('Lateral (cm)')
        #axes[i][j].yaxis.label('Axial (cm)')
        #el = patches.Ellipse((cc[0],cc[1]), 2*cc[2], 2*cc[3], fill=False,edgecolor='red')
        #el.set_linestyle('--')
        #axes[i][j].add_patch(el)
        count=count+1

plt.show()
#fig = plt.gcf()
fig.savefig('Image_Grid_no_markers.png')
```
